# Theoritical Questions

1. What is hypothesis testing in statistics?
   -> Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves testing an            assumption (null hypothesis) to determine if there is enough evidence to reject it.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
   -> The null hypothesis (H₀) assumes there is no effect or difference in the population. The alternative hypothesis (H₁) claims there is a                  significant effect or difference.

3. What is the significance level in hypothesis testing, and why is it important?
   -> The significance level (α) is the probability of rejecting the null hypothesis when it is actually true, commonly set at 0.05. It determines the        threshold for statistical significance and helps control the risk of a Type I error.

4. What does a P-value represent in hypothesis testing?
   -> The P-value represents the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis is           true. A smaller P-value indicates stronger evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
   -> If the P-value is less than the significance level (α), you reject the null hypothesis. If the P-value is greater than α, you fail to reject the        null hypothesis.

```python
6. What are Type 1 and Type 2 errors in hypothesis testing?
   -> Type I error: rejecting the null hypothesis when it is actually true (a false positive).
      Type II error: failing to reject the null hypothesis when it is actually false (a false negative).
```

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   -> A one-tailed test checks for an effect in one specific direction (greater or less than).
      A two-tailed test checks for an effect in both directions (different from, not equal to).

8. What is the Z-test, and when is it used in hypothesis testing?
   -> A Z-test is used to determine if there is a significant difference between sample and population means when the population variance is known. It        is typically used for large sample sizes (n ≥ 30).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
   -> The Z-score is calculated as $Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}$, where $\bar{X}$ is the sample mean, $\mu$ is the population mean,           $\sigma$ is the population standard deviation, and $n$ is the sample size.
      It represents how many standard deviations the sample mean is from the population mean.

```python
10. What is the T-distribution, and when should it be used instead of the normal distribution?
    -> The T-distribution is a probability distribution used when the sample size is small and the population standard deviation is unknown.
       It is preferred over the normal distribution when $n < 30$ or when dealing with sample data.
```

11. What is the difference between a Z-test and a T-test?
    -> A Z-test is used when the population variance is known and the sample size is large.
       A T-test is used when the population variance is unknown and the sample size is small.

12. What is the T-test, and how is it used in hypothesis testing?
    -> A T-test compares the sample mean to a population mean to determine if they differ significantly when population variance is unknown.
       It’s used with small samples to test hypotheses about the mean.

13. What is the relationship between Z-test and T-test in hypothesis testing?
    -> Both Z-test and T-test assess if sample data differ from a population or another sample.
       Z-test applies with known population variance and large samples; T-test is used when variance is unknown or sample size is small.

```python
14. What is a confidence interval, and how is it used to interpret statistical results?
    -> A confidence interval is a range of values that likely contains the true population parameter with a given confidence level (e.g., 95%).
       It helps estimate the precision and reliability of a sample statistic in hypothesis testing.
```

```python
15. What is the margin of error, and how does it affect the confidence interval?
    -> The margin of error is the maximum expected difference between the sample estimate and the true population value.
       A larger margin of error widens the confidence interval, reducing precision.
```

```python
16. How is Bayes' Theorem used in statistics, and what is its significance?
    -> Bayes' Theorem updates the probability of a hypothesis based on new evidence or data.
       It’s significant for combining prior knowledge with observed data to improve decision-making.
```

```python
17. What is the Chi-square distribution, and when is it used?
    -> The Chi-square distribution is used to analyze categorical data by comparing observed and expected frequencies.
       It’s commonly used in tests of independence and goodness-of-fit.
```

```python
18. What is the Chi-square goodness of fit test, and how is it applied?
    -> The Chi-square goodness of fit test checks if observed categorical data matches expected distribution frequencies.
       It’s applied by comparing observed counts to expected counts to see if differences are due to chance.
```

```python
19. What is the F-distribution, and when is it used in hypothesis testing?
    -> The F-distribution is used to compare variances between two or more groups in hypothesis testing.
       It’s commonly applied in ANOVA and tests for equality of variances.
```

```python
20. What is an ANOVA test, and what are its assumptions?
    -> ANOVA (Analysis of Variance) tests if there are significant differences between means of three or more groups.
       Assumptions: normality, independence, and equal variances (homogeneity) across groups.
```

```python
21. What are the different types of ANOVA tests?
    -> One-way ANOVA: compares means across one factor with multiple groups.
    -> Two-way ANOVA: compares means across two factors and their interaction.
    -> Repeated Measures ANOVA: compares means within the same subjects over time or conditions.
```

```python
22. What is the F-test, and how does it relate to hypothesis testing?
    -> The F-test compares variances to determine if group means are significantly different in hypothesis testing.
       It uses the F-distribution to assess if observed differences are due to chance.
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
#code
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")
    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.6
print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.
#code
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    
    # Calculate Z-score
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    
    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z)))
    
    # Print results
    print(f"Sample mean: {sample_mean:.4f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    # Decision
    if p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

# Example usage:
population_mean = 50
population_std = 8
sample_data = np.array([52, 48, 49, 51, 50, 53, 47, 54, 49, 50])

z_test(sample_data, population_mean, population_std)
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
#code
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # standard error of the mean
    
    # Calculate the margin of error using t-distribution
    h = sem * stats.t.ppf((1 + confidence) / 2, n - 1)
    
    lower = mean - h
    upper = mean + h
    
    print(f"Sample mean = {mean:.3f}")
    print(f"{int(confidence*100)}% Confidence interval: ({lower:.3f}, {upper:.3f})")
    print("Interpretation: We are confident that the true population mean lies within this interval.")
    
    return lower, upper

# Example dataset
data = np.array([12, 15, 14, 16, 13, 15, 17, 14, 16, 15])
confidence_interval(data)
```

```python
'''24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.'''
#code
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(size=1000, mean=0, std=1):
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)
    
    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # excess kurtosis (normal = 0)
    
    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (excess): {data_kurtosis:.4f}")
    
    return data_skewness, data_kurtosis

# Example usage
analyze_distribution()
```

```python
'''1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.'''
#code
import numpy as np
from scipy.stats import norm

def z_test_sample_mean(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f"Sample mean = {sample_mean:.2f}")
    print(f"Z-score = {z_score:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Hypothesis testing decision
    if p_value < alpha:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

# Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=52, scale=10, size=30)
population_mean = 50
population_std = 10

z_test_sample_mean(sample_data, population_mean, population_std)
```

```python
#3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
#code
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    
    # Calculate Z-score
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    
    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Display results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Decision
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (significant difference).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (no significant difference).")

# Example usage
np.random.seed(123)
sample_data = np.random.normal(loc=51, scale=10, size=50)  # Simulated sample
population_mean = 50
population_std = 10  # Known population standard deviation

one_sample_z_test(sample_data, population_mean, population_std)
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
#code
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Simulate sample data
np.random.seed(0)
sample = np.random.normal(loc=52, scale=8, size=40)
population_mean = 50
population_std = 8  # known population standard deviation

# Step 2: Perform Z-test
n = len(sample)
sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 3: Define critical Z-values for alpha = 0.05 (two-tailed)
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

# Step 4: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print("Conclusion:", "Reject H0" if abs(z_score) > z_critical else "Fail to reject H0")

# Step 5: Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (α/2)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3)

# Plot z-score
plt.axvline(z_score, color='green', linestyle='--', label=f'Z = {z_score:.2f}')
plt.axvline(-z_critical, color='black', linestyle='--', label=f'Critical Z = ±{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle='--')

plt.title('Two-Tailed Z-test Visualization')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
#code
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0=50, mu_1=53, sigma=10, n=30, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for a one-tailed test (right tail)
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu_0 + z_critical * se

    # Beta: Probability of Type II error (fail to reject H0 when H1 is true)
    beta = norm.cdf((x_critical - mu_1) / se)

    # x values for plotting
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    y_H0 = norm.pdf(x, loc=mu_0, scale=se)
    y_H1 = norm.pdf(x, loc=mu_1, scale=se)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_H0, label='H₀ Distribution (mean = {:.0f})'.format(mu_0), color='blue')
    plt.plot(x, y_H1, label='H₁ Distribution (mean = {:.0f})'.format(mu_1), color='green')

    # Shade Type I error region (alpha)
    plt.fill_between(x, y_H0, where=(x > x_critical), color='red', alpha=0.4, label='Type I Error (α)')

    # Shade Type II error region (beta)
    plt.fill_between(x, y_H1, where=(x < x_critical), color='orange', alpha=0.4, label='Type II Error (β)')

    plt.axvline(x_critical, color='black', linestyle='--', label='Critical Value')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Output numeric results
    print(f"Critical value (Xc): {x_critical:.2f}")
    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage
visualize_type1_type2_errors()
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results.
#code
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):
    # Perform independent t-test
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welch's t-test

    # Print results
    print(f"Sample 1 Mean: {np.mean(sample1):.2f}")
    print(f"Sample 2 Mean: {np.mean(sample2):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret result
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (means are significantly different).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (no significant difference).")

# Example usage
np.random.seed(0)
sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=53, scale=5, size=30)

independent_t_test(sample1, sample2)
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results.
#code
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test_with_visualization(before, after, alpha=0.05):
    # Perform paired t-test
    t_stat, p_value = ttest_rel(before, after)

    # Print statistical results
    print(f"Mean Before: {np.mean(before):.2f}")
    print(f"Mean After: {np.mean(after):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret result
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (significant change).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (no significant change).")

    # Visualization: Line plot of paired values
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    for i in range(len(before)):
        plt.plot(['Before', 'After'], [before[i], after[i]], marker='o', color='skyblue')
    plt.title("Paired Observations Comparison")
    plt.ylabel("Value")

    # Visualization: Histogram of differences
    differences = after - before
    plt.subplot(1, 2, 2)
    plt.hist(differences, bins=10, color='orange', edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='--')
    plt.title("Histogram of Differences (After - Before)")
    plt.xlabel("Difference")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

# Example usage
np.random.seed(42)
before = np.random.normal(loc=70, scale=5, size=30)
after = before + np.random.normal(loc=-2, scale=3, size=30)  # Simulate slight decrease

paired_t_test_with_visualization(before, after)
```

```python
#8.Simulate data and perform both Z-test and T-test, then compare the results using Python.
#code
import numpy as np
from scipy.stats import norm, t

def z_test(sample, population_mean, population_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    return z_stat, p_value

def t_test(sample, population_mean, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    se = sample_std / np.sqrt(n)
    t_stat = (sample_mean - population_mean) / se
    p_value = 2 * (1 - t.cdf(abs(t_stat), df=n-1))
    return t_stat, p_value

# Simulate sample data
np.random.seed(0)
population_mean = 50
population_std = 10  # Known population std for Z-test
sample = np.random.normal(loc=52, scale=population_std, size=30)

# Perform Z-test
z_stat, z_p = z_test(sample, population_mean, population_std)

# Perform T-test
t_stat, t_p = t_test(sample, population_mean)

# Print results
print(f"Z-test: Z = {z_stat:.4f}, p-value = {z_p:.4f}")
print(f"T-test: t = {t_stat:.4f}, p-value = {t_p:.4f}")

# Interpretation
alpha = 0.05
print("\nInterpretation:")
if z_p < alpha:
    print("- Z-test: Reject the null hypothesis.")
else:
    print("- Z-test: Fail to reject the null hypothesis.")

if t_p < alpha:
    print("- T-test: Reject the null hypothesis.")
else:
    print("- T-test: Fail to reject the null hypothesis.")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
#code
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean = {mean:.3f}")
    print(f"{int(confidence*100)}% Confidence Interval = ({lower_bound:.3f}, {upper_bound:.3f})")

    return lower_bound, upper_bound

# Example usage:
import numpy as np
np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=15, size=40)
confidence_interval(sample_data)
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
#code
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_crit * std_err

    print(f"Sample size: {n}")
    print(f"Confidence level: {confidence*100}%")
    print(f"Margin of Error: {moe:.4f}")

    return moe

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=40)
margin_of_error(sample_data, confidence=0.95)
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
#code
def bayes_theorem(prior_disease, sensitivity, false_positive_rate):
    # Prior probabilities
    prior_no_disease = 1 - prior_disease

    # Likelihoods
    p_positive_given_disease = sensitivity
    p_positive_given_no_disease = false_positive_rate

    # Total probability of positive test
    p_positive = (p_positive_given_disease * prior_disease) + (p_positive_given_no_disease * prior_no_disease)

    # Posterior probability (Bayes' Theorem)
    posterior = (p_positive_given_disease * prior_disease) / p_positive

    print(f"Prior Probability of Disease: {prior_disease:.2%}")
    print(f"Probability of Positive Test: {p_positive:.2%}")
    print(f"Posterior Probability of Disease given Positive Test: {posterior:.2%}")

    return posterior

# Example parameters
prior = 0.01  # 1% prevalence
sensitivity = 0.99  # 99% true positive rate
false_positive_rate = 0.05  # 5% false positive rate

bayes_theorem(prior, sensitivity, false_positive_rate)
```

```python
#12.Perform a Chi-square test for independence between two categorical variables in Python.
#code
import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table (observed frequencies)
# Rows: Gender (Male, Female)
# Columns: Preference (Product A, Product B, Product C)
data = np.array([[30, 10, 20],
                 [15, 25, 20]])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("Conclusion: Reject null hypothesis - variables are dependent.")
else:
    print("Conclusion: Fail to reject null hypothesis - variables are independent.")
```

```python
'''13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data'''
#code
import numpy as np

def calculate_expected_frequencies(observed):
    # Calculate row totals and column totals
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    # Calculate expected frequencies using formula:
    # expected[i,j] = (row_total[i] * col_total[j]) / total
    expected = (row_totals @ col_totals) / total
    return expected

# Example observed data (contingency table)
observed_data = np.array([[30, 10, 20],
                          [15, 25, 20]])

expected_freq = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_freq)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
#code
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (e.g., counts in categories)
observed = np.array([50, 30, 20])

# Expected frequencies (e.g., from a theoretical distribution)
# Make sure expected sums to the same total as observed
expected_probs = np.array([0.4, 0.4, 0.2])  # Expected proportions
expected = expected_probs * observed.sum()

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Observed frequencies: {observed}")
print(f"Expected frequencies: {expected.astype(int)}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — observed distribution differs from expected.")
else:
    print("Conclusion: Fail to reject the null hypothesis — observed distribution fits expected.")
```
